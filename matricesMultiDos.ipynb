```python
def menu():
    print("\n--- Menú ---")
    print("1. Ingresar nueva persona")
    print("2. Mostrar todos los datos")
    print("3. Filtrar por DNI")
    print("4. Salir")
    
def ingresar_persona():
    nombre = input("Ingresa el nombre: ")
    apellido = input("Ingresa el apellido: ")
    dni = input("Ingresa el DNI: ")
    
    # Ingreso de teléfonos
    telefonos_input = input("Ingresa los teléfonos separados por comas: ")
    telefonos = telefonos_input.split(',')
    
    # Ingreso de hijos
    hijos_input = input("Ingresa los nombres de los hijos separados por comas: ")
    hijos = hijos_input.split(',')
    
    persona = [nombre, apellido, dni, telefonos, hijos]
    return persona

def mostrar_datos(personas):
    if personas:
        print("\nMatriz de datos ingresados:")
        print(personas)  # Mostrar la matriz completa

        print("\nDetalles de cada persona:")
        for persona in personas:
            nombre, apellido, dni, telefonos, hijos = persona
            print(f"{nombre} {apellido}, DNI: {dni}")
            print(f"Teléfonos: {', '.join(telefonos)}")
            print(f"Hijos: {', '.join(hijos)}")
            print(f"Cantidad de teléfonos: {len(telefonos)}")
            print(f"Cantidad de hijos: {len(hijos)}\n")
    else:
        print("\nNo hay personas ingresadas aún.")

def filtrar_por_dni(personas):
    dni_buscado = input("Ingresa el DNI para filtrar: ")
    encontrado = False
    for persona in personas:
        if persona[2] == dni_buscado:
            nombre, apellido, dni, telefonos, hijos = persona
            print(f"\nDatos de {nombre} {apellido}:")
            print(f"DNI: {dni}")
            print(f"Teléfonos: {', '.join(telefonos)}")
            print(f"Hijos: {', '.join(hijos)}")
            print(f"Cantidad de teléfonos: {len(telefonos)}")
            print(f"Cantidad de hijos: {len(hijos)}\n")
            encontrado = True
            break
    if not encontrado:
        print("\nNo se encontró una persona con ese DNI.")

# Llamada directa a la función main
personas = []
while True:
    menu()
    opcion = input("Elige una opción: ")
    
    if opcion == '1':
        persona = ingresar_persona()
        personas.append(persona)
    elif opcion == '2':
        mostrar_datos(personas)
    elif opcion == '3':
        filtrar_por_dni(personas)
    elif opcion == '4':
        print("¡Hasta luego!")
        break
    else:
        print("Opción no válida, por favor ingresa una opción correcta.")
```
